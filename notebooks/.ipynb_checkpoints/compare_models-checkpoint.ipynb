```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import load_model
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score
from data_loader import DataLoader
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
loader = DataLoader(data_dir='/content/Images/Images', img_size = (32,32))

X_train, X_val, X_test_1C, y_train, y_val, y_test, classes = loader.load_data()

X_test_3C = np.repeat(X_test_1C, 3, axis=-1)
```

```python
X_test_1C.shape
```

```python
X_test_3C.shape
```

```python
MODEL_PATHS = {
    "CNN":"/content/drive/MyDrive/models/cnn_model.keras",
    "MobileNetV2":"/content/drive/MyDrive/models/mobilenetv2_best.keras",
    "MobileNetV2FineTuned":"/content/drive/MyDrive/models/mobilenetv2_finetuned_best.keras"
}
```

```python
models = {name: load_model(path) for name, path in MODEL_PATHS.items()}
```

```python
models
```

```python
for model_name, model in models.items():
  if model_name == "CNN":
    preds = model.predict(X_test_1C)
    preds = np.argmax(preds, axis = 1)

    test_loss, test_acc = model.evaluate(X_test_1C, y_test)

    class_labels = classes.classes_
    cm = confusion_matrix(y_test, preds)

    plt.figure(figsize=(15, 12))
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        xticklabels=class_labels,
        yticklabels=class_labels
    )
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Confusion Matrix: {model_name} with Accuracy: {test_acc*100:.2f}%')
    plt.show()
  else:
    preds = model.predict(X_test_3C)
    preds = np.argmax(preds, axis = 1)

    test_loss, test_acc = model.evaluate(X_test_3C, y_test)

    class_labels = classes.classes_
    cm = confusion_matrix(y_test, preds)

    plt.figure(figsize=(15, 12))
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        xticklabels=class_labels,
        yticklabels=class_labels
    )
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Confusion Matrix: {model_name} with Accuracy: {test_acc*100:.2f}%')
    plt.show()


```
